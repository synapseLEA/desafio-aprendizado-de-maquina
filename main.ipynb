```python
feature = ['glucose_postprandial', 'bmi']
x = df[feature]
x.head()
```

```python
y = df.diagnosed_diabetes
display(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
model = SVC(kernel = 'rbf', random_state = 0)
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

y_pred = model.predict(x_test)

cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from matplotlib.colors import ListedColormap
plt.figure(figsize=(6,5))
X1, X2 = x_test[:,0], x_test[:,1]
xx, yy = np.meshgrid(np.linspace(X1.min()-1, X1.max()+1, 200),
                     np.linspace(X2.min()-1, X2.max()+1, 200))
plt.contourf(xx, yy, model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape),
             alpha=0.3, cmap=ListedColormap(['red','green']))
plt.scatter(X1, X2, c=y_test, cmap=ListedColormap(['black','white']))
plt.xlabel('glucose_postprandial')
plt.ylabel('bmi')
plt.title('SVM com Kernel RBF')
plt.show()
```
